```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate data
x = np.linspace(0, 2 * np.pi, 100)  # 0 to 2Ï€, 100 points
y = np.sin(x)  # Sine of each point
```

```python
# Plot the data
plt.plot(x, y, label='Sine Wave')
plt.title('Sine Wave Example')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
